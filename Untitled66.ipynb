```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.colors import LinearSegmentedColormap

# 
plt.style.use('seaborn-v0_8-white')
plt.rcParams['font.sans-serif'] = ['Arial', 'SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False

def generate_figure_1():
    # 13
    fig = plt.figure(figsize=(18, 6), constrained_layout=True)
    gs = fig.add_gridspec(1, 3, width_ratios=[1, 1, 1.2]) # 3

    # ==========================================
    #  A: Turing Instability (Reaction-Diffusion)
    # ==========================================
    ax1 = fig.add_subplot(gs[0, 0])

    #  Turing  ()
    x = np.linspace(0, 10, 100)
    y = np.linspace(0, 10, 100)
    X, Y = np.meshgrid(x, y)

    # 
    Z_turing = (np.sin(X*1.5) * np.sin(Y*1.5) +
                0.5 * np.sin(X*3 + Y*0.5) +
                0.5 * np.cos(X*0.5 + Y*3))

    # 
    cmap_rd = plt.get_cmap('viridis')
    im1 = ax1.imshow(Z_turing, cmap=cmap_rd, extent=[0, 10, 0, 10], origin='lower')

    # 
    ax1.set_title("A. Turing Instability\n(Reaction-Diffusion)", fontsize=14, fontweight='bold')
    ax1.set_xlabel("Space $x$")
    ax1.set_ylabel("Space $y$")
    ax1.set_xticks([])
    ax1.set_yticks([])

    # 
    ax1.text(0.5, -0.15, "Mechanism: Diffusion + Reaction\nResult: Spatial Patterns (Spots/Stripes)",
             transform=ax1.transAxes, ha='center', va='top', fontsize=11, color='#333')

    # ==========================================
    #  B: Chemotaxis (Keller-Segel)
    # ==========================================
    ax2 = fig.add_subplot(gs[0, 1])

    #  Keller-Segel  ()
    def gaussian(x, y, x0, y0, sigma):
        return np.exp(-((x-x0)**2 + (y-y0)**2) / (2 * sigma**2))

    Z_ks = (gaussian(X, Y, 3, 3, 0.8) +
            gaussian(X, Y, 7, 8, 1.0) * 1.2 +
            gaussian(X, Y, 8, 2, 0.6) * 0.8)

    #  - 
    cmap_ks = plt.get_cmap('magma')
    im2 = ax2.imshow(Z_ks, cmap=cmap_ks, extent=[0, 10, 0, 10], origin='lower')

    #  ( Advection/Drift)
    # 
    dy, dx = np.gradient(Z_ks)
    skip = 8
    ax2.quiver(X[::skip, ::skip], Y[::skip, ::skip],
               dx[::skip, ::skip], dy[::skip, ::skip],
               color='white', alpha=0.6, scale=15, width=0.005)

    # 
    ax2.set_title("B. Chemotaxis\n(Keller-Segel)", fontsize=14, fontweight='bold')
    ax2.set_xlabel("Space $x$")
    ax2.set_yticks([])
    ax2.set_xticks([])

    # 
    ax2.text(0.5, -0.15, "Mechanism: Advection along $\\nabla c$\nResult: Aggregation / Blow-up",
             transform=ax2.transAxes, ha='center', va='top', fontsize=11, color='#333')

    # ==========================================
    #  C: Hybrid TME Model (Diagram)
    # ==========================================
    ax3 = fig.add_subplot(gs[0, 2])
    ax3.set_xlim(0, 10)
    ax3.set_ylim(0, 10)
    ax3.axis('off') # 

    # 
    color_cells = '#e74c3c'  # 
    color_signal = '#3498db' # 
    color_env = '#95a5a6'    # 

    # 

    # 1. : Motile Cells (PDE)
    rect_cells = patches.Rectangle((2, 7), 6, 2, linewidth=2, edgecolor=color_cells, facecolor='none', linestyle='--')
    ax3.add_patch(rect_cells)
    ax3.text(5, 8, "Motile Population (S, R)\n(PDE: Diffusion + Chemotaxis)", ha='center', va='center', fontsize=10, color=color_cells, fontweight='bold')

    # 2. : Signal (PDE)
    rect_signal = patches.Rectangle((2, 4), 6, 2, linewidth=2, edgecolor=color_signal, facecolor='none')
    ax3.add_patch(rect_signal)
    ax3.text(5, 5, "Diffusive Signal $c$\n(PDE: Smooths the input)", ha='center', va='center', fontsize=10, color=color_signal, fontweight='bold')

    # 3. : Environment (ODE)
    rect_env = patches.Rectangle((2, 1), 6, 2, linewidth=2, edgecolor=color_env, facecolor='#f0f0f0')
    ax3.add_patch(rect_env)
    ax3.text(5, 2, "Microenv. State $A$\n(ODE: Pointwise Switching)", ha='center', va='center', fontsize=10, color='#555', fontweight='bold')

    # ---  ---

    # Arrow 1: A produces c (Source)
    # 
    ax3.annotate("", xy=(5, 4), xytext=(5, 3),
                 arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax3.text(5.2, 3.5, "Source", fontsize=9, ha='left')

    # Arrow 2: c guides Cells (Chemotaxis)
    # 
    ax3.annotate("", xy=(5, 7), xytext=(5, 6),
                 arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax3.text(5.2, 6.5, "Senses $\\nabla c$", fontsize=9, ha='left')

    # Feedback Visualization ()
    #  (Optional Feedback Loop)
    ax3.annotate("", xy=(1.8, 2), xytext=(1.8, 8),
                 arrowprops=dict(arrowstyle="->", lw=1.5, color='gray', linestyle='--'))
    ax3.text(0.8, 5, "Feedback?", fontsize=9, ha='center', rotation=90, color='gray')

    # 
    ax3.set_title("C. Hybrid TME Framework\n(Motivation)", fontsize=14, fontweight='bold')

    # 
    ax3.text(0.5, -0.15, "Gap: Signal $c$ acts as proxy\nfor non-diffusive env $A$\nto define drift $\\nabla c$.",
             transform=ax3.transAxes, ha='center', va='top', fontsize=11, color='#333')

    # 
    plt.suptitle("Comparison of Pattern Formation Mechanisms & Proposed Hybrid Model", fontsize=16, y=1.05)

    # 
    # plt.savefig('figure1_mechanism.png', dpi=300, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    generate_figure_1()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def draw_figure_2_architecture():
    # 
    fig, ax = plt.subplots(figsize=(12, 10), constrained_layout=True)
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.axis('off')  # 

    # ==========================================
    # 
    # ==========================================
    # 
    color_pop = '#FFEBEE'    #  (Population)
    border_pop = '#C62828'   # 
    color_sig = '#E3F2FD'    #  (Signal)
    border_sig = '#1565C0'   # 
    color_env = '#F5F5F5'    #  (Environment)
    border_env = '#616161'   # 

    box_width = 6
    box_height = 1.8
    center_x = 5

    # ==========================================
    #  (Layers)
    # ==========================================

    # 1. : Population Layer (PDE)
    # -------------------------------------------------
    y_top = 7.5
    box_top = patches.FancyBboxPatch(
        (center_x - box_width/2, y_top), box_width, box_height,
        boxstyle="round,pad=0.1", fc=color_pop, ec=border_pop, lw=2
    )
    ax.add_patch(box_top)

    # 
    ax.text(center_x, y_top + box_height - 0.4, "Top Layer: Motile Population (PDE)",
            ha='center', va='center', fontsize=12, fontweight='bold', color=border_pop)
    ax.text(center_x, y_top + box_height/2, r"Variables: $S(x,t), R(x,t)$",
            ha='center', va='center', fontsize=14)
    ax.text(center_x, y_top + 0.4, r"Mechanisms: Diffusion $\Delta \mathbf{u}$ + Chemotaxis $\nabla \cdot (S\chi \nabla c)$",
            ha='center', va='center', fontsize=11, color='#444')

    # 2. : Signal Layer (PDE) - The "Proxy"
    # -------------------------------------------------
    y_mid = 4.5
    box_mid = patches.FancyBboxPatch(
        (center_x - box_width/2, y_mid), box_width, box_height,
        boxstyle="round,pad=0.1", fc=color_sig, ec=border_sig, lw=2
    )
    ax.add_patch(box_mid)

    # 
    ax.text(center_x, y_mid + box_height - 0.4, "Middle Layer: Diffusive Signal (PDE)",
            ha='center', va='center', fontsize=12, fontweight='bold', color=border_sig)
    ax.text(center_x, y_mid + box_height/2, r"Variable: $c(x,t)$",
            ha='center', va='center', fontsize=14)
    ax.text(center_x, y_mid + 0.4, "Role: Diffusive Proxy for A\n(Makes $\\nabla c$ well-defined)",
            ha='center', va='center', fontsize=11, color='#444', style='italic')

    # 3. : Environment Layer (ODE)
    # -------------------------------------------------
    y_bot = 1.5
    box_bot = patches.FancyBboxPatch(
        (center_x - box_width/2, y_bot), box_width, box_height,
        boxstyle="round,pad=0.1", fc=color_env, ec=border_env, lw=2
    )
    ax.add_patch(box_bot)

    # 
    ax.text(center_x, y_bot + box_height - 0.4, "Bottom Layer: Microenvironment (ODE)",
            ha='center', va='center', fontsize=12, fontweight='bold', color=border_env)
    ax.text(center_x, y_bot + box_height/2, r"Variable: $A(x,t)$",
            ha='center', va='center', fontsize=14)
    ax.text(center_x, y_bot + 0.4, "Mechanisms: Pointwise Switching / Transitions\n(NO Diffusion $\\to$ No Spatial Gradient)",
            ha='center', va='center', fontsize=11, color='#444')

    # ==========================================
    #  (Arrows)
    # ==========================================

    #  A -> A (Solid Arrow: Production)
    # -------------------------------------
    ax.annotate("", xy=(center_x, y_mid), xytext=(center_x, y_bot + box_height),
                arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax.text(center_x + 0.1, (y_mid + y_bot + box_height)/2,
            "Source\n$c_t = ... + \kappa A$", fontsize=10, ha='left', va='center')

    #  c -> S (Solid Arrow: Sensing)
    # -------------------------------------
    ax.annotate("", xy=(center_x, y_top), xytext=(center_x, y_mid + box_height),
                arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax.text(center_x + 0.1, (y_top + y_mid + box_height)/2,
            "Chemotactic Drift\nSenses $\\nabla c$", fontsize=10, ha='left', va='center')

    # : One-way Flow
    # -------------------------------------
    ax.annotate("One-way Coupling\n(Damped)", xy=(center_x - 3.2, 5.5), xytext=(center_x - 3.2, 5.5),
                ha='center', va='center', fontsize=12, fontweight='bold', color='#2E7D32', rotation=90)
    ax.annotate("", xy=(center_x - 2.8, 8.5), xytext=(center_x - 2.8, 2.5),
                arrowprops=dict(arrowstyle="->", lw=3, color='#2E7D32', alpha=0.5))

    # : Feedback Loop (Dashed)
    # -------------------------------------
    # 
    ax.annotate("", xy=(center_x + 2.8, 2.5), xytext=(center_x + 2.8, 8.5),
                arrowprops=dict(arrowstyle="->", lw=2, color='#E65100', linestyle='--', alpha=0.8))

    ax.text(center_x + 3.2, 5.5, "Two-way Feedback\n(Destabilizing?)",
            ha='center', va='center', fontsize=12, fontweight='bold', color='#E65100', rotation=-90)

    #  (S -> c, S -> A)
    #  Top 
    ax.annotate("", xy=(center_x + 3.0, 5.5), xytext=(center_x + 3.0, 7.8),
                 arrowprops=dict(arrowstyle="-", lw=1, color='#E65100', linestyle='--'))
    #  Middle
    ax.annotate("", xy=(center_x + box_width/2, 5.5), xytext=(center_x + 3.0, 5.5),
                 arrowprops=dict(arrowstyle="->", lw=1, color='#E65100', linestyle='--'))
    ax.text(center_x + 3.5, 6.0, "Cells produce/degrade $c$\n(Feedback)", fontsize=9, color='#E65100')

    # ==========================================
    #  (The "Why")
    # ==========================================
    # 

    # 
    ax_inset = fig.add_axes([0.05, 0.45, 0.15, 0.15]) # [left, bottom, width, height]
    ax_inset.set_title("Why introduce $c$?", fontsize=9, fontweight='bold')
    ax_inset.set_xticks([])
    ax_inset.set_yticks([])

    # 
    x_small = np.linspace(0, 10, 50)
    # 1. Step function (A) - non-diffusive
    y_A = np.where(x_small < 5, 0, 1)
    ax_inset.plot(x_small, y_A, color='gray', label='$A$ (Step)', lw=1.5)

    # 2. Smooth function (c) - diffusive
    y_c = 1 / (1 + np.exp(-2 * (x_small - 5))) # Sigmoid
    ax_inset.plot(x_small, y_c, color='blue', label='$c$ (Smooth)', lw=1.5, linestyle='--')

    ax_inset.legend(fontsize=7, loc='upper left')
    ax_inset.text(5, -0.4, r"$\nabla A$ is $\delta$-func!" + "\n" + r"$\nabla c$ is bounded.",
                  transform=ax_inset.transData, ha='center', va='top', fontsize=8, color='red')

    # 
    plt.suptitle("Figure 2: Hybrid PDE-ODE Model Architecture & Signal Layer Rationale", fontsize=16, fontweight='bold', y=0.98)

    plt.show()

if __name__ == "__main__":
    draw_figure_2_architecture()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def draw_figure_3_dichotomy():
    # 
    plt.style.use('seaborn-v0_8-white')
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['axes.unicode_minus'] = False

    #  ()
    col_pop = '#d32f2f'   # Red: Population (u)
    col_sig = '#1976d2'   # Blue: Signal (c)
    col_env = '#757575'   # Gray: Environment (A)
    col_bg_unstable = '#fff3e0' # 

    # : 2 (Scenario 1 vs Scenario 2) x 3 (Flow, Spectrum, Result)
    # height_ratios 
    fig = plt.figure(figsize=(14, 10), constrained_layout=True)
    gs = fig.add_gridspec(3, 2, height_ratios=[1.2, 1.5, 1])

    # =================================================================
    # Column 1:  (One-way / Pattern Suppression)
    # =================================================================

    # --- Row 1: Logic Flow ---
    ax_flow_1 = fig.add_subplot(gs[0, 0])
    ax_flow_1.set_xlim(0, 10); ax_flow_1.set_ylim(0, 10); ax_flow_1.axis('off')
    ax_flow_1.set_title("A. One-way Damped Coupling (C4)\n(Pattern Suppression)", fontsize=14, fontweight='bold', color='#2e7d32')

    # 
    # Top: Pop
    rect_u = patches.FancyBboxPatch((3, 7.5), 4, 1.5, boxstyle="round,pad=0.1", fc='white', ec=col_pop, lw=2)
    ax_flow_1.add_patch(rect_u)
    ax_flow_1.text(5, 8.25, r"Population $\mathbf{u}=(S,R)$", ha='center', va='center', color=col_pop, fontweight='bold')

    # Mid: Signal
    rect_c = patches.FancyBboxPatch((3, 4.5), 4, 1.5, boxstyle="round,pad=0.1", fc='white', ec=col_sig, lw=2)
    ax_flow_1.add_patch(rect_c)
    ax_flow_1.text(5, 5.25, r"Signal $c$ (Diffusive)", ha='center', va='center', color=col_sig, fontweight='bold')

    # Bot: Env
    rect_a = patches.FancyBboxPatch((3, 1.5), 4, 1.5, boxstyle="round,pad=0.1", fc='#f5f5f5', ec=col_env, lw=2)
    ax_flow_1.add_patch(rect_a)
    ax_flow_1.text(5, 2.25, r"Env State $A$ (ODE)", ha='center', va='center', color=col_env, fontweight='bold')

    # Arrows (One Way Up)
    ax_flow_1.annotate("", xy=(5, 4.5), xytext=(5, 3.0), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_1.text(5.2, 3.75, "Source", fontsize=9)

    ax_flow_1.annotate("", xy=(5, 7.5), xytext=(5, 6.0), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_1.text(5.2, 6.75, "Chemotaxis\n(No Feedback)", fontsize=9)

    # Math Note
    ax_flow_1.text(8, 5, r"Jacobian $\mathbf{J}$ is"+"\nBlock-Triangular", ha='center', va='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray', boxstyle='round'))

    # --- Row 2: Spectrum (Complex Plane) ---
    ax_spec_1 = fig.add_subplot(gs[1, 0])
    ax_spec_1.set_title("Linear Stability Spectrum", fontsize=11)
    ax_spec_1.axvline(0, color='black', linestyle='--', alpha=0.5) # Imaginary Axis
    ax_spec_1.axhline(0, color='black', linestyle='-', alpha=0.3)
    ax_spec_1.set_xlim(-5, 2); ax_spec_1.set_ylim(-3, 3)
    ax_spec_1.set_xlabel(r"Re$(\lambda)$"); ax_spec_1.set_ylabel(r"Im$(\lambda)$")

    #  ()
    # Pop eigenvalues (random stable)
    evals_pop = -np.random.rand(15)*2 - 0.5 + 1j * (np.random.rand(15)-0.5)*4
    ax_spec_1.scatter(evals_pop.real, evals_pop.imag, c=col_pop, marker='x', label='Pop Modes')

    # Signal eigenvalues (stable real)
    evals_c = -np.random.rand(5) - 2
    ax_spec_1.scatter(evals_c, np.zeros_like(evals_c), c=col_sig, marker='o', label='Signal Modes')

    ax_spec_1.text(-4, 2, "Stable\nRe$(\lambda) < 0$", color='#2e7d32', fontweight='bold')
    ax_spec_1.legend(loc='lower left', fontsize=8)

    # --- Row 3: Spatial Result ---
    ax_spat_1 = fig.add_subplot(gs[2, 0])
    ax_spat_1.set_title("Spatial Outcome: Homogeneous", fontsize=11)
    x = np.linspace(0, 10, 100)
    # Flat lines with slight decay noise
    ax_spat_1.plot(x, np.ones_like(x)*1.5, color=col_pop, lw=2, label=r'$u(x)$')
    ax_spat_1.plot(x, np.ones_like(x)*1.0, color=col_sig, lw=2, linestyle='--', label=r'$c(x)$')
    ax_spat_1.set_ylim(0, 2.5)
    ax_spat_1.legend(loc='upper right', fontsize=8)
    ax_spat_1.set_xlabel("Space $x$")
    ax_spat_1.text(5, 0.5, "No Pattern", ha='center', fontsize=12, color='#2e7d32', fontweight='bold')


    # =================================================================
    # Column 2:  (Two-way / Pattern Recovery)
    # =================================================================

    # --- Row 1: Logic Flow ---
    ax_flow_2 = fig.add_subplot(gs[0, 1])
    ax_flow_2.set_xlim(0, 10); ax_flow_2.set_ylim(0, 10); ax_flow_2.axis('off')
    ax_flow_2.set_title("B. Two-way Feedback (C5)\n(Pattern Recovery)", fontsize=14, fontweight='bold', color='#c62828')

    # Draw Boxes (Same positions)
    # Top
    rect_u2 = patches.FancyBboxPatch((3, 7.5), 4, 1.5, boxstyle="round,pad=0.1", fc='white', ec=col_pop, lw=2)
    ax_flow_2.add_patch(rect_u2)
    ax_flow_2.text(5, 8.25, r"Population $\mathbf{u}=(S,R)$", ha='center', va='center', color=col_pop, fontweight='bold')
    # Mid
    rect_c2 = patches.FancyBboxPatch((3, 4.5), 4, 1.5, boxstyle="round,pad=0.1", fc='white', ec=col_sig, lw=2)
    ax_flow_2.add_patch(rect_c2)
    ax_flow_2.text(5, 5.25, r"Signal $c$ (Diffusive)", ha='center', va='center', color=col_sig, fontweight='bold')
    # Bot
    rect_a2 = patches.FancyBboxPatch((3, 1.5), 4, 1.5, boxstyle="round,pad=0.1", fc='#f5f5f5', ec=col_env, lw=2)
    ax_flow_2.add_patch(rect_a2)
    ax_flow_2.text(5, 2.25, r"Env State $A$ (ODE)", ha='center', va='center', color=col_env, fontweight='bold')

    # Arrows (Two Way Feedback)
    # Mid <-> Bot
    ax_flow_2.annotate("", xy=(4.5, 4.5), xytext=(4.5, 3.0), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_2.annotate("", xy=(5.5, 3.0), xytext=(5.5, 4.5), arrowprops=dict(arrowstyle="->", lw=2, color='black', linestyle='--'))
    ax_flow_2.text(5.6, 3.75, "Feedback", fontsize=9, color='red')

    # Top <-> Mid
    ax_flow_2.annotate("", xy=(4.5, 7.5), xytext=(4.5, 6.0), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_2.annotate("", xy=(5.5, 6.0), xytext=(5.5, 7.5), arrowprops=dict(arrowstyle="->", lw=2, color='black', linestyle='--'))
    ax_flow_2.text(5.6, 6.75, "Prod/Degrad.", fontsize=9, color='red')

    # Math Note
    ax_flow_2.text(8.2, 5, "Effective\nCross-Diffusion", ha='center', va='center', fontsize=10, bbox=dict(facecolor=col_bg_unstable, alpha=1, edgecolor='red', boxstyle='round'))

    # --- Row 2: Spectrum ---
    ax_spec_2 = fig.add_subplot(gs[1, 1])
    ax_spec_2.set_title("Instability: Eigenvalues Cross Axis", fontsize=11)
    # Highlight unstable region
    rect_unstable = patches.Rectangle((0, -3), 2, 6, facecolor=col_bg_unstable, alpha=0.5)
    ax_spec_2.add_patch(rect_unstable)

    ax_spec_2.axvline(0, color='black', linestyle='--', alpha=0.5)
    ax_spec_2.axhline(0, color='black', linestyle='-', alpha=0.3)
    ax_spec_2.set_xlim(-5, 2); ax_spec_2.set_ylim(-3, 3)
    ax_spec_2.set_xlabel(r"Re$(\lambda)$"); ax_spec_2.set_yticks([])

    # Stable modes
    evals_stable = -np.random.rand(10)*3 - 1 + 1j * (np.random.rand(10)-0.5)*4
    ax_spec_2.scatter(evals_stable.real, evals_stable.imag, c='gray', alpha=0.5, label='Stable Modes')

    # Unstable Turing branch (Parabolic path crossing 0)
    k = np.linspace(0, 1, 10)
    # Simulate a dispersion relation curve crossing 0
    re_lambda = -1.5 + 4 * k - 2 * k**2 # quadratic opening left but crossing right
    im_lambda = np.zeros_like(k)

    # Plot the path
    ax_spec_2.plot(re_lambda, im_lambda, 'k--', alpha=0.3)
    # Highlight the unstable mode
    unstable_pts = re_lambda > 0
    ax_spec_2.scatter(re_lambda[unstable_pts], im_lambda[unstable_pts], c='red', s=80, marker='*', label='Unstable Modes')
    ax_spec_2.scatter(re_lambda[~unstable_pts], im_lambda[~unstable_pts], c='black', s=10)

    ax_spec_2.text(0.2, 2, "Unstable!\nRe$(\lambda) > 0$", color='#c62828', fontweight='bold')
    ax_spec_2.legend(loc='lower left', fontsize=8)

    # --- Row 3: Spatial Result ---
    ax_spat_2 = fig.add_subplot(gs[2, 1])
    ax_spat_2.set_title("Spatial Outcome: Turing Patterns", fontsize=11)

    # Sine waves
    y_pat_u = 1.5 + 0.4 * np.cos(3 * x)
    y_pat_c = 1.0 + 0.3 * np.cos(3 * x) # In phase or anti-phase depending on kinetics, assume in phase for demo

    ax_spat_2.plot(x, y_pat_u, color=col_pop, lw=2, label=r'$u(x)$')
    ax_spat_2.plot(x, y_pat_c, color=col_sig, lw=2, linestyle='--', label=r'$c(x)$')
    ax_spat_2.set_ylim(0, 2.5)
    ax_spat_2.set_xlabel("Space $x$")
    ax_spat_2.set_yticks([])
    ax_spat_2.text(5, 0.2, "Pattern Formation", ha='center', fontsize=12, color='#c62828', fontweight='bold')

    # Main Title
    plt.suptitle("Main Analytical Results C4 vs C5", fontsize=16, y=1.02)

    # Save
    # plt.savefig('figure3_stability_dichotomy.png', dpi=300, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    draw_figure_3_dichotomy()
```

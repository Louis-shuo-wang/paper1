```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.colors import LinearSegmentedColormap

# 
plt.style.use('seaborn-v0_8-white')
plt.rcParams['font.sans-serif'] = ['Arial', 'SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False

def generate_figure_1():
    # 13
    fig = plt.figure(figsize=(18, 6), constrained_layout=True)
    gs = fig.add_gridspec(1, 3, width_ratios=[1, 1, 1.2]) # 3

    # ==========================================
    #  A: Turing Instability (Reaction-Diffusion)
    # ==========================================
    ax1 = fig.add_subplot(gs[0, 0])

    #  Turing  ()
    x = np.linspace(0, 10, 100)
    y = np.linspace(0, 10, 100)
    X, Y = np.meshgrid(x, y)

    # 
    Z_turing = (np.sin(X*1.5) * np.sin(Y*1.5) +
                0.5 * np.sin(X*3 + Y*0.5) +
                0.5 * np.cos(X*0.5 + Y*3))

    # 
    cmap_rd = plt.get_cmap('viridis')
    im1 = ax1.imshow(Z_turing, cmap=cmap_rd, extent=[0, 10, 0, 10], origin='lower')

    # 
    ax1.set_title("A. Turing Instability\n(Reaction-Diffusion)", fontsize=20, fontweight='bold')
    ax1.set_xlabel("Space $x$", fontsize=16)
    ax1.set_ylabel("Space $y$", fontsize=16)
    ax1.set_xticks([])
    ax1.set_yticks([])

    # 
    ax1.text(0.5, -0.15, "Mechanism: Diffusion + Reaction\nResult: Spatial Patterns (Spots/Stripes)",
             transform=ax1.transAxes, ha='center', va='top', fontsize=20, color='#333')

    # ==========================================
    #  B: Chemotaxis (Keller-Segel)
    # ==========================================
    ax2 = fig.add_subplot(gs[0, 1])

    #  Keller-Segel  ()
    def gaussian(x, y, x0, y0, sigma):
        return np.exp(-((x-x0)**2 + (y-y0)**2) / (2 * sigma**2))

    Z_ks = (gaussian(X, Y, 3, 3, 0.8) +
            gaussian(X, Y, 7, 8, 1.0) * 1.2 +
            gaussian(X, Y, 8, 2, 0.6) * 0.8)

    #  - 
    cmap_ks = plt.get_cmap('magma')
    im2 = ax2.imshow(Z_ks, cmap=cmap_ks, extent=[0, 10, 0, 10], origin='lower')

    #  ( Advection/Drift)
    # 
    dy, dx = np.gradient(Z_ks)
    skip = 8
    ax2.quiver(X[::skip, ::skip], Y[::skip, ::skip],
               dx[::skip, ::skip], dy[::skip, ::skip],
               color='white', alpha=0.6, scale=15, width=0.005)

    # 
    ax2.set_title("B. Chemotaxis\n(Keller-Segel)", fontsize=20, fontweight='bold')
    ax2.set_xlabel("Space $x$", fontsize=16)
    ax2.set_yticks([])
    ax2.set_xticks([])

    # 
    ax2.text(0.5, -0.15, "Mechanism: Advection along $\\nabla c$\nResult: Aggregation / Blow-up",
             transform=ax2.transAxes, ha='center', va='top', fontsize=20, color='#333')

    # ==========================================
    #  C: Hybrid TME Model (Diagram)
    # ==========================================
    ax3 = fig.add_subplot(gs[0, 2])
    ax3.set_xlim(0, 10)
    ax3.set_ylim(0, 10)
    ax3.axis('off') # 

    # 
    color_cells = '#e74c3c'  # 
    color_signal = '#3498db' # 
    color_env = '#95a5a6'    # 

    # 

    # 1. : Motile Cells (PDE)
    rect_cells = patches.Rectangle((2, 7), 6, 2, linewidth=2, edgecolor=color_cells, facecolor='none', linestyle='--')
    ax3.add_patch(rect_cells)
    ax3.text(5, 8, "Motile Population (S, R)\n(PDE: Diffusion + Chemotaxis)", ha='center', va='center', fontsize=20, color=color_cells, fontweight='bold')

    # 2. : Signal (PDE)
    rect_signal = patches.Rectangle((2, 4), 6, 2, linewidth=2, edgecolor=color_signal, facecolor='none')
    ax3.add_patch(rect_signal)
    ax3.text(5, 5, "Diffusive Signal $c$\n(PDE: Smooths the input)", ha='center', va='center', fontsize=20, color=color_signal, fontweight='bold')

    # 3. : Environment (ODE)
    rect_env = patches.Rectangle((2, 1), 6, 2, linewidth=2, edgecolor=color_env, facecolor='#f0f0f0')
    ax3.add_patch(rect_env)
    ax3.text(5, 2, "Microenv. State $A$\n(ODE: Pointwise Switching)", ha='center', va='center', fontsize=20, color='#555', fontweight='bold')

    # ---  ---

    # Arrow 1: A produces c (Source)
    # 
    ax3.annotate("", xy=(5, 4), xytext=(5, 3),
                 arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax3.text(5.2, 3.5, "Source", fontsize=20, ha='left')

    # Arrow 2: c guides Cells (Chemotaxis)
    # 
    ax3.annotate("", xy=(5, 7), xytext=(5, 6),
                 arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax3.text(5.2, 6.5, "Senses $\\nabla c$", fontsize=20, ha='left')

    # Feedback Visualization ()
    #  (Optional Feedback Loop)
    ax3.annotate("", xy=(1.8, 2), xytext=(1.8, 8),
                 arrowprops=dict(arrowstyle="->", lw=1.5, color='gray', linestyle='--'))
    ax3.text(0.8, 5, "Feedback?", fontsize=20, ha='center', rotation=90, color='gray')

    # 
    ax3.set_title("C. Hybrid TME Framework\n(Motivation)", fontsize=20, fontweight='bold')
    # 
    ax3.text(0.5, -0.15, "Gap: Signal $c$ acts as proxy\nfor non-diffusive env $A$\nto define drift $\\nabla c$.",
             transform=ax3.transAxes, ha='center', va='top', fontsize=20, color='#333')

    # 
    plt.suptitle("Comparison of Pattern Formation Mechanisms & Proposed Hybrid Model", fontsize=24, y=1.05)

    # 
    # plt.savefig('figure1_mechanism.png', dpi=600, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    generate_figure_1()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def draw_figure_3_dichotomy():
    # =========================
    # 
    # =========================
    fig_width = 7.0   # inches: 7 inch  DPI  one-column
    fig_height = 10.0
    base_font = 11         # 
    panel_title_font = 18  # A/B 
    box_text_font = 10     # 
    arrow_text_font = 11   # 
    note_font = 10         # 
    suptitle_font = 16

    plt.style.use('seaborn-v0_8-white')
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['axes.unicode_minus'] = False
    plt.rcParams.update({'font.size': base_font})

    # 
    col_pop = '#d32f2f'   # Red: Population (u)
    col_sig = '#1976d2'   # Blue: Signal (c)
    col_env = '#757575'   # Gray: Environment (A)
    col_bg_unstable = '#fff3e0' # 

    # 
    fig = plt.figure(figsize=(fig_width, fig_height), constrained_layout=True)
    gs = fig.add_gridspec(3, 2, height_ratios=[1.2, 1.5, 1])

    # ===  ===
    box_x = 3.0
    box_width = 6.5
    box_height = 1.6
    vspace = 1.5  # 
    top_y = 8.2
    mid_y = top_y - (box_height + vspace)
    bot_y = mid_y - (box_height + vspace)
    note_gap = 0.6          #  box 
    note_x = box_x + box_width + note_gap
    # =================================================================
    # Column 1:  (One-way / Pattern Suppression)
    # =================================================================

    # --- Row 1: Logic Flow ---
    ax_flow_1 = fig.add_subplot(gs[0, 0])
    ax_flow_1.set_xlim(0, 10); ax_flow_1.set_ylim(0, 10); ax_flow_1.axis('off')
    ax_flow_1.set_title("A. One-way Damped Coupling \n(Pattern Suppression)",
                        fontsize=panel_title_font, fontweight='bold', color='#2e7d32')

    # Top: Pop
    rect_u = patches.FancyBboxPatch((box_x, top_y), box_width, box_height,
                                    boxstyle="round,pad=0.1", fc='white', ec=col_pop, lw=2)
    ax_flow_1.add_patch(rect_u)
    ax_flow_1.text(box_x + box_width / 2, top_y + box_height / 2,
                   r"Population $\mathbf{u}=(S,R)$", ha='center', va='center',
                   color=col_pop, fontweight='bold', fontsize=box_text_font)

    # Mid: Signal
    rect_c = patches.FancyBboxPatch((box_x, mid_y), box_width, box_height,
                                    boxstyle="round,pad=0.1", fc='white', ec=col_sig, lw=2)
    ax_flow_1.add_patch(rect_c)
    ax_flow_1.text(box_x + box_width / 2, mid_y + box_height / 2,
                   r"Signal $c$ (Diffusive)", ha='center', va='center',
                   color=col_sig, fontweight='bold', fontsize=box_text_font)

    # Bot: Env
    rect_a = patches.FancyBboxPatch((box_x, bot_y), box_width, box_height,
                                    boxstyle="round,pad=0.1", fc='#f5f5f5', ec=col_env, lw=2)
    ax_flow_1.add_patch(rect_a)
    ax_flow_1.text(box_x + box_width / 2, bot_y + box_height / 2,
                   r"Env State $A$ (ODE)", ha='center', va='center',
                   color=col_env, fontweight='bold', fontsize=box_text_font)

    # Arrows (One Way Up)  
    ax_flow_1.annotate("", xy=(5, mid_y), xytext=(5, bot_y + box_height), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_1.text(5.2, bot_y + box_height + (mid_y - (bot_y + box_height)) / 2, "Source",
                   fontsize=arrow_text_font, ha='left', va='center')

    ax_flow_1.annotate("", xy=(5, top_y), xytext=(5, mid_y + box_height), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_1.text(5.2, mid_y + box_height + (top_y - (mid_y + box_height)) / 2, "Chemotaxis\n(No Feedback)",
                   fontsize=arrow_text_font, ha='left', va='center')

    # Math Note
    ax_flow_1.text(note_x, mid_y + box_height/2,
               r"Jacobian $\mathbf{J}$ is" + "\nBlock-Triangular",
               ha='left', va='center', fontsize=note_font,
               bbox=dict(facecolor='white', alpha=0.95,
                         edgecolor='gray', boxstyle='round'))
    
    # --- Row 2: Spectrum (Complex Plane) ---
    ax_spec_1 = fig.add_subplot(gs[1, 0])
    ax_spec_1.set_title("Linear Stability Spectrum", fontsize=panel_title_font-6)
    ax_spec_1.axvline(0, color='black', linestyle='--', alpha=0.5) # Imaginary Axis
    ax_spec_1.axhline(0, color='black', linestyle='-', alpha=0.3)
    ax_spec_1.set_xlim(-5, 2); ax_spec_1.set_ylim(-3, 3)
    ax_spec_1.set_xlabel(r"Re$(\lambda)$", fontsize=base_font)
    ax_spec_1.set_ylabel(r"Im$(\lambda)$", fontsize=base_font)
    ax_spec_1.tick_params(labelsize=base_font)

    # Pop eigenvalues (random stable)
    evals_pop = -np.random.rand(15)*2 - 0.5 + 1j * (np.random.rand(15)-0.5)*4
    ax_spec_1.scatter(evals_pop.real, evals_pop.imag, c=col_pop, marker='x', label='Pop Modes')

    # Signal eigenvalues (stable real)
    evals_c = -np.random.rand(5) - 2
    ax_spec_1.scatter(evals_c, np.zeros_like(evals_c), c=col_sig, marker='o', label='Signal Modes')

    ax_spec_1.text(-4, 2, "Stable\nRe$(\lambda) < 0$", color='#2e7d32', fontweight='bold', fontsize=note_font)
    ax_spec_1.legend(loc='lower left', fontsize=base_font)

    # --- Row 3: Spatial Result ---
    ax_spat_1 = fig.add_subplot(gs[2, 0])
    ax_spat_1.set_title("Spatial Outcome: Homogeneous", fontsize=panel_title_font-6)
    x = np.linspace(0, 10, 100)
    ax_spat_1.plot(x, np.ones_like(x)*1.5, color=col_pop, lw=2, label=r'$u(x)$')
    ax_spat_1.plot(x, np.ones_like(x)*1.0, color=col_sig, lw=2, linestyle='--', label=r'$c(x)$')
    ax_spat_1.set_ylim(0, 2.5)
    ax_spat_1.legend(loc='upper right', fontsize=base_font)
    ax_spat_1.set_xlabel("Space $x$", fontsize=base_font)
    ax_spat_1.tick_params(labelsize=base_font)
    ax_spat_1.text(5, 0.5, "No Pattern", ha='center', fontsize=box_text_font, color='#2e7d32', fontweight='bold')

    # =================================================================
    # Column 2:  (Two-way / Pattern Recovery)
    # =================================================================

    # --- Row 1: Logic Flow ---
    ax_flow_2 = fig.add_subplot(gs[0, 1])
    ax_flow_2.set_xlim(0, 10); ax_flow_2.set_ylim(0, 10); ax_flow_2.axis('off')
    ax_flow_2.set_title("B. Two-way Feedback \n(Pattern Recovery)",
                        fontsize=panel_title_font, fontweight='bold', color='#c62828')

    # Draw Boxes (same horizontal positions, use the computed vertical positions)
    rect_u2 = patches.FancyBboxPatch((box_x, top_y), box_width, box_height, boxstyle="round,pad=0.1", fc='white', ec=col_pop, lw=2)
    ax_flow_2.add_patch(rect_u2)
    ax_flow_2.text(box_x + box_width / 2, top_y + box_height / 2,
                   r"Population $\mathbf{u}=(S,R)$", ha='center', va='center',
                   color=col_pop, fontweight='bold', fontsize=box_text_font)

    rect_c2 = patches.FancyBboxPatch((box_x, mid_y), box_width, box_height, boxstyle="round,pad=0.1", fc='white', ec=col_sig, lw=2)
    ax_flow_2.add_patch(rect_c2)
    ax_flow_2.text(box_x + box_width / 2, mid_y + box_height / 2,
                   r"Signal $c$ (Diffusive)", ha='center', va='center',
                   color=col_sig, fontweight='bold', fontsize=box_text_font)

    rect_a2 = patches.FancyBboxPatch((box_x, bot_y), box_width, box_height, boxstyle="round,pad=0.1", fc='#f5f5f5', ec=col_env, lw=2)
    ax_flow_2.add_patch(rect_a2)
    ax_flow_2.text(box_x + box_width / 2, bot_y + box_height / 2,
                   r"Env State $A$ (ODE)", ha='center', va='center',
                   color=col_env, fontweight='bold', fontsize=box_text_font)

    # Arrows (Two Way Feedback)  
    # Mid <-> Bot
    ax_flow_2.annotate("", xy=(4.5, mid_y), xytext=(4.5, bot_y + box_height), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_2.annotate("", xy=(5.5, bot_y + box_height), xytext=(5.5, mid_y), arrowprops=dict(arrowstyle="->", lw=2, color='black', linestyle='--'))
    ax_flow_2.text(5.6, bot_y + box_height + (mid_y - (bot_y + box_height)) / 2, "Feedback",
                   fontsize=arrow_text_font, color='red', ha='left', va='center')

    # Top <-> Mid
    ax_flow_2.annotate("", xy=(4.5, top_y), xytext=(4.5, mid_y + box_height), arrowprops=dict(arrowstyle="->", lw=2, color='black'))
    ax_flow_2.annotate("", xy=(5.5, mid_y + box_height), xytext=(5.5, top_y), arrowprops=dict(arrowstyle="->", lw=2, color='black', linestyle='--'))
    ax_flow_2.text(5.6, mid_y + box_height + (top_y - (mid_y + box_height)) / 2, "Prod/Degrad.",
                   fontsize=arrow_text_font, color='red', ha='left', va='center')

    # Math Note
    ax_flow_2.text(note_x, mid_y + box_height/2,
               "Effective\nCross-Diffusion",
               ha='left', va='center', fontsize=note_font,
               bbox=dict(facecolor=col_bg_unstable, alpha=1.0,
                         edgecolor='red', boxstyle='round'))

    # --- Row 2: Spectrum ---
    ax_spec_2 = fig.add_subplot(gs[1, 1])
    ax_spec_2.set_title("Instability: Eigenvalues Cross Axis", fontsize=panel_title_font-6)
    rect_unstable = patches.Rectangle((0, -3), 2, 6, facecolor=col_bg_unstable, alpha=0.5)
    ax_spec_2.add_patch(rect_unstable)

    ax_spec_2.axvline(0, color='black', linestyle='--', alpha=0.5)
    ax_spec_2.axhline(0, color='black', linestyle='-', alpha=0.3)
    ax_spec_2.set_xlim(-5, 2); ax_spec_2.set_ylim(-3, 3)
    ax_spec_2.set_xlabel(r"Re$(\lambda)$", fontsize=base_font)
    ax_spec_2.set_yticks([])
    ax_spec_2.tick_params(labelsize=base_font)

    # Stable modes
    evals_stable = -np.random.rand(10)*3 - 1 + 1j * (np.random.rand(10)-0.5)*4
    ax_spec_2.scatter(evals_stable.real, evals_stable.imag, c='gray', alpha=0.5, label='Stable Modes')

    # Unstable Turing branch (Parabolic path crossing 0)
    k = np.linspace(0, 1, 10)
    re_lambda = -1.5 + 4 * k - 2 * k**2
    im_lambda = np.zeros_like(k)
    ax_spec_2.plot(re_lambda, im_lambda, 'k--', alpha=0.3)
    unstable_pts = re_lambda > 0
    ax_spec_2.scatter(re_lambda[unstable_pts], im_lambda[unstable_pts], c='red', s=80, marker='*', label='Unstable Modes')
    ax_spec_2.scatter(re_lambda[~unstable_pts], im_lambda[~unstable_pts], c='black', s=10)

    ax_spec_2.text(0.2, 2, "Unstable!\nRe$(\lambda) > 0$", color='#c62828', fontweight='bold', fontsize=note_font)
    ax_spec_2.legend(loc='upper left', fontsize=base_font)

    # --- Row 3: Spatial Result ---
    ax_spat_2 = fig.add_subplot(gs[2, 1])
    ax_spat_2.set_title("Spatial Outcome: Turing Patterns", fontsize=panel_title_font-6)

    x = np.linspace(0, 10, 100)
    y_pat_u = 1.5 + 0.4 * np.cos(3 * x)
    y_pat_c = 1.0 + 0.3 * np.cos(3 * x)

    ax_spat_2.plot(x, y_pat_u, color=col_pop, lw=2, label=r'$u(x)$')
    ax_spat_2.plot(x, y_pat_c, color=col_sig, lw=2, linestyle='--', label=r'$c(x)$')
    ax_spat_2.set_ylim(0, 2.5)
    ax_spat_2.set_xlabel("Space $x$", fontsize=base_font)
    ax_spat_2.set_yticks([])
    ax_spat_2.tick_params(labelsize=base_font)
    ax_spat_2.text(5, 0.2, "Pattern Formation", ha='center', fontsize=box_text_font, color='#c62828', fontweight='bold')
    ax_spat_2.legend(loc='upper left', fontsize=base_font)

    # Main Title
    plt.suptitle("Main Analytical Results", fontsize=suptitle_font, y=1.02)

    # 
    #  PDF DPI PNG
    # plt.savefig('figure3_stability_dichotomy.pdf', bbox_inches='tight')
    # plt.savefig('figure3_stability_dichotomy.png', dpi=600, bbox_inches='tight')

    plt.show()

if __name__ == "__main__":
    draw_figure_3_dichotomy()
```
